# Spark Basics Test

This notebook tests basic PySpark functionality including:
- SparkSession creation
- DataFrame creation from Python data
- Basic transformations (select, filter, withColumn)
- Aggregations

## 1. Create SparkSession

```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import col, lit, when, avg, sum, count

# Create SparkSession
spark = SparkSession.builder \
    .appName("Spark Basics Test") \
    .config("spark.sql.adaptive.enabled", "true") \
    .getOrCreate()

print(f"Spark version: {spark.version}")
print(f"Spark app name: {spark.sparkContext.appName}")
print("\nSparkSession created successfully!")
```

## 2. Create DataFrame from Python Data

```python
# Sample employee data
employees = [
    (1, "Alice", "Engineering", 75000),
    (2, "Bob", "Engineering", 80000),
    (3, "Charlie", "Sales", 65000),
    (4, "Diana", "Sales", 70000),
    (5, "Eve", "Marketing", 60000),
    (6, "Frank", "Engineering", 90000),
    (7, "Grace", "Marketing", 55000),
    (8, "Henry", "Sales", 72000),
]

columns = ["id", "name", "department", "salary"]

# Create DataFrame
df = spark.createDataFrame(employees, columns)

print("Employee DataFrame created!")
print(f"Number of rows: {df.count()}")
print(f"Number of columns: {len(df.columns)}")
```

## 3. Examine DataFrame Schema and Contents

```python
# Print schema
print("DataFrame Schema:")
df.printSchema()

# Show data
print("\nDataFrame Contents:")
df.show()
```

## 4. Basic Operations: Select and Filter

```python
# Select specific columns
print("Select name and department:")
df.select("name", "department").show()

# Filter rows
print("\nEmployees with salary > 70000:")
df.filter(col("salary") > 70000).show()

# Filter by department
print("\nEngineering department only:")
df.filter(col("department") == "Engineering").show()
```

## 5. Add New Columns with withColumn

```python
# Add a bonus column (10% of salary)
df_with_bonus = df.withColumn("bonus", col("salary") * 0.10)

# Add salary level column
df_with_level = df_with_bonus.withColumn(
    "salary_level",
    when(col("salary") >= 80000, "High")
    .when(col("salary") >= 65000, "Medium")
    .otherwise("Low")
)

print("DataFrame with bonus and salary level:")
df_with_level.show()
```

## 6. GroupBy and Aggregations

```python
# Group by department and calculate statistics
print("Department Statistics:")
department_stats = df.groupBy("department").agg(
    count("*").alias("employee_count"),
    avg("salary").alias("avg_salary"),
    sum("salary").alias("total_salary")
)

department_stats.show()

# Order by average salary
print("\nDepartments ordered by average salary (descending):")
department_stats.orderBy(col("avg_salary").desc()).show()
```

## 7. Stop SparkSession

```python
spark.stop()
print("SparkSession stopped successfully!")
print("\nSpark basics test completed!")
```
