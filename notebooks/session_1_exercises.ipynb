## Exercise 2: Safe Type Conversion


```python
"""
Create a function that safely converts values from CSV strings to their
appropriate Python types. Handle edge cases gracefully.
"""


def convert_value(value, target_type):
    """
    Safely convert a string value to the target type.

    Args:
        value: String value to convert (may be empty or invalid)
        target_type: One of 'int', 'float', 'bool', 'str'

    Returns:
        Converted value, or None if conversion fails

    Examples:
        convert_value("42", "int") → 42
        convert_value("", "int") → None
        convert_value("true", "bool") → True
        convert_value("3.14", "float") → 3.14
    """
   
    try:
        if target_type=="int":
            return int(value)

        elif target_type == "float":
            return float(value)

        elif target_type == "bool":
            if value.lower() == "true":
                return True
            elif value.lower() == "false":
                return False
            else:
                return None

        elif target_type == "str":
            return value

        else:
            return None
    except:
        return None
values = ["42", "", "3.14", "true", "false", "hello"]
types = ["int", "int", "float", "bool", "bool", "str"]

results = []

for i in range(len(values)):
    results.append(convert_value(values[i], types[i]))

print(results)

```

Test your code



```python
print(convert_value("42", "int"))        # Should return 42

print(convert_value("", "int"))          # Should return None

print(convert_value("3.14", "float"))    # Should return 3.14

print(convert_value("true", "bool"))     # Should return True

print(convert_value("false", "bool"))    # Should return False

print(convert_value("hello", "str"))     # Should return "hello"
```

## Exercise 3: File Path Parser


```python
"""
Create a function that parses a file path and extracts information using
string methods and slicing.
"""
import re


def parse_file_path(path):
    """
    Parse a file path and extract components.

    Args:
        path: File path string like "/data/raw/customers_2024_01_15.csv"

    Returns:
        dict with keys:
        - directory: Parent directory path
        - filename: File name without extension
        - extension: File extension (without dot)
        - date: Date string if present in filename (YYYY_MM_DD format), else None

    Example:
        parse_file_path("/data/raw/customers_2024_01_15.csv")
        → {
            "directory": "/data/raw",
            "filename": "customers_2024_01_15",
            "extension": "csv",
            "date": "2024_01_15"
        }
    """
    if "/" in path:
        directory, filename_full = path.rsplit("/", 1)
    else:
        directory = ""
        filename_full = path

    filename, extension = filename_full.rsplit(".", 1)

    match = re.search(r'\d{4}_\d{2}_\d{2}', filename)
    date = match.group() if match else None

    return {
        "directory": directory,
        "filename": filename,
        "extension": extension,
        "date": date
    }
```

Test your code


```python
print(parse_file_path("/data/raw/customers_2024_01_15.csv")) # should return {'directory': '/data/raw', 'filename': 'customers_2024_01_15', 'extension': 'csv', 'date': '2024_01_15'}
print(parse_file_path("/home/user/reports/summary.json")) # should return {'directory': '/home/user/reports', 'filename': 'summary', 'extension': 'json', 'date': None}
print(parse_file_path("simple.txt")) # should return {'directory': '', 'filename': 'simple', 'extension': 'txt', 'date': None}
```

## Exercise 4: Status Router


```python
"""
Using match/case (Python 3.10+) or if/elif/else, create a function that
routes pipeline events to appropriate handlers based on event type.
"""


def route_event(event):
    """
    Route a pipeline event to the appropriate action.

    Args:
        event: Tuple of (event_type, event_data)
            - ("start", pipeline_name) → "Starting pipeline: {name}"
            - ("complete", record_count) → "Completed: {count} records"
            - ("error", error_message) → "ERROR: {message}"
            - ("warning", warning_message) → "WARNING: {message}"
            - ("metric", (name, value)) → "Metric {name}={value}"
            - Any other → "Unknown event: {type}"

    Returns:
        Formatted message string
    """
    event_type = event[0]
    event_data = event[1]
    
    if event_type == "start":
        return f"Starting pipeline: {event_data}"
    elif event_type == "complete":
        return f"Completed: {event_data} record"
    elif event_type=="error":
        return f"ERROR: {event_data}"
    elif  event_type == "metric":
          name, value = event_data
          return f"metric {name}={value}"
    else: 
        return f"unknown_type: {event_type}"
         
```

Test your code


```python
print(route_event(("start", "daily_etl"))) # Starting pipeline: daily_etl

print(route_event(("complete", 1500))) # Completed: 1500 records

print(route_event(("error", "Connection timeout"))) # ERROR: Connection timeout

print(route_event(("metric", ("cpu_usage", 75)))) # Metric cpu_usage=75

print(route_event(("unknown_type", "data"))) # Unknown event: unknown_type
```
